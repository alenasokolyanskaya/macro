### Based on Problem Set 3. Question 2. General Equilibrium with Bequest

As an extension of the model solved in the Problem Set 3 of the General Equilibrium model, I decided to change the initial wealth that agents receive when they are young at the first period and introduce bequest that young generation gets from the old. 

So the agents need to optimize their utility, such that in the first period of their life they get bequest, and in the second period when they are old, they have to leave bequest for the next generation.

An altruistic bequest motive makes finitely lived individuals lived for a finite number of periods to maximize the utility of the entire dynasty. 

Utility function with bequest: we view second-period consumption and bequests as a composite good.

$ U = \alpha ln (c) + (1- \alpha)[ \beta  ln (c') + (1- \beta) ln (b')]$

s.t. 

$c + a = wh + b$

$ c' = a (1+r) + w' h' - b'$

FOCs:

assests: [a] $a = (1-\alpha) (w'h + b) $

consumption when young: [c] $c = \alpha (wh + b - a)$

consumption when old: [c'] $c' = \beta (1- \alpha) (1+r) (wh+b-a)$

bequest decision when old: [b'] $ (1-\beta) (1-\alpha) (wh+b-a) (1+r)$

By manipulating values of $\alpha$ and $\beta$, we change how agents care about the consumption when being young and old, and how much when being old they care about next generation, while making decision on the amount of the bequest.

$0<\alpha<1$, $0<\beta<1$: The preference parameter $\alpha$ captures intertemporal preferences and $\beta $ captures the relative importance of second-period consumption and bequest.

So, for instance, if I increase the value of $\alpha$ the agent cares more about consuming when young, and if I decrease $\beta$ then the agent when being old cares more about leaving larger bequest for the next generation.

In the first part I present the results for very altruistic generation with $\beta=0.1$ and $\alpha = 0.51$.
And in the second case I assume that the agents are less altruistic so I increase $\beta = 0.9$

In the first case (very altruistic case) decision path of the bequest decided when being old, increases rapidly after approximately 5 periods, and the agents seek to increase the welfareof the entire dynasty, reaching the level of the bequest at the level of approximately 3, slightly fluctuating over the time (T=100) (see Fig.1).

Since we assume that our world in this model is very altruistic, old generation cares a lot about the wealth of the young generation. So we can see 
that decision path of the young is much higher than the consumption of old (Fig.2). 

When I changed $\beta =0.9$, the level of the bequest decreased to approximately 0.2 (Fig.3).

As a result of making our agents caring less about the next generation, we got that the agents prefer to consume more when being old (Fig.4).


## [1] Very altruistic case: $\beta=0.1$ $\alpha = 0.51$

```python
import numpy as np
from numpy import *
from scipy.optimize import *
from itertools import product
import matplotlib.pyplot as plt

random.seed(10)
#DISTRIBUTION OBJECT
eps=np.random.choice([-0.05,0.05],size=100,p=[0.5,0.5])
eps = np.array(eps)

# Solve the system of equations
solution1=[]

def decision(b0, eps):
    r=0.01 #interest rate
    nu=4 #productivity
    w0 = nu
    w1 = nu + eps
    beta=0.1
    alpha = 0.51
    h = 1

    a = (1-alpha)*(w1*h + b0)
    c0 = alpha*(w0*h + b0 -a)
    c1 = beta * a * (1+r)
    b = (1-beta)*a* (1+r)
    
    return a, c0, c1, b

#simulate
Sol = [0]*100
for i in range(100):
    Sol[i] = [0]*4

#def values for t = 0
Sol[0][3]=0.0125

for t in range(1, 100):
    Sol[t] = decision(Sol[t-1][3], eps[t])
```

```python
#Sol
```

```python
Sol = np.array(Sol)
```

```python
# bequest that is left each period by old
plt.plot(range(100), Sol[:,3])
plt.xlabel('time')
plt.ylabel('b when being old')
plt.title('Fig.1: Decision path on the bequest left by old generation')
```

```python
plt.plot(range(100), Sol[:,1], label='c young')
plt.plot(range(100), Sol[:,2], label='c old')
plt.legend()
plt.xlabel('time')
plt.ylabel('consumption')
plt.title('Fig.2: Decision path on the consumption by young and old')
```
